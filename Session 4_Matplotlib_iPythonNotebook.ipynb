This Code shows how to plot a graph using the Ipyhton notebook to code for the programme to plot the selected data on the graph using matplotlib. Matplotlib is the programme used to plot the actual graph.

First we are going to import numpy as np so that it is easy to use when using certain functions of numpy. We are also going to import matplotlib as plt so it is easy to use in the porgramme.

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

The next line tells the programme to put all the plots of the graph in this notebook rather than an external file

```python
%matplotlib inline
```

Now, we are going to create an array which gives values from 0 and $$2*pi$$ with 100 divisions between them, which we have named 'x'. Then we are going to apply the sin function to this, named as 'y', to and plot the graph of x values with y.

```python
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('A graph showing sin(x) of all values of x from 0 and 2$pi$')
```

The above graph shows the plotting of x and y=sinx

Now we are going to create a new array of numbers from 0 to 10 with 100 divisions between them.

```python
x = np.linspace(0, 10, 100)
```

We are going to apply two functions to this array which are sinx and cosx which will then be plotted into the graph together so we can compare both of them. The axis titles and legend along with the title of the graph will be shown aswell.

```python
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1, label = "sin(x)")
plt.plot(x,y2, label = "cos(x)")
plt.xlabel('x (radians)')
plt.ylabel('function of x')
plt.title('The sin and cos functions of x')
plt.legend(loc="lower left")
```

The above shows the graph of the sine and cosine functions of the above array of x-values plotted together on one graph. Now we are going to use this technique for our own set of values by uploading a file consisting data from an experiment involving the X-ray diffraction due to certain chemical compounds.

First, we are going to upload the files onto python by using the code listed below. To show our data, we are also going to 'print' the values - which can be good to identify any extreme values in the results.

The first set of data we are going to upload is going to be Lithium Flouride.

```python
xdata,xcounts = np.loadtxt("Lif-30kV.csv", delimiter=',', unpack = True)
print "our 2xtheta for Lithium Chloride = ", xdata
print "counts per second Lithium Chloride = ", xcounts
```

Now, we are going to upload the data for Sodium Chloride and the X-ray diffraction due to this compound.

```python
ydata,ycounts = np.loadtxt("NaCl-30kV.csv", delimiter=',', unpack=True)
print "our 2xtheta for Sodium Chloride = " , ydata
print "counts per second for Sodium Chloride = ", ycounts
```

We are now goiung to plot the results for both the chemical compounds and the X-ray diffraction due to them on a single graph so we can compare and contrast between the. We can use different colours on the graph and styles to show the difference and make the results clear.

```python
plt.plot(xdata, xcounts,'r.', label = "Lithium Flouride")
plt.plot(ydata, ycounts, 'c.', label = "Sodium Chloride")
plt.xlabel('Angle 2*theta')
plt.ylabel('Counts per second')
plt.title('X-ray diffraction data for a crystal of each compound')
plt.legend(loc = "upper right")
plt.grid(True)
plt.savefig('X-ray Diffraction.pdf')
plt.savefig('X-ray Diffraction.png', dpi=300)
```

There you can see a graph of the results and the data uploaded for each of the compounds
